### Imports

```python
# beautifuSoup to handle the HTML data.
from bs4 import BeautifulSoup 

# To make connection to the web page
import urllib.request
```

```python
# specify the url
targeturl =  'https://www.valueresearchonline.com/stocks/stockselector/default.asp?index=46'
```

```python
# query the website and return the html to the variable "targetpage"
targetpage = urllib.request.urlopen(targeturl)
# parse the html using beautiful soup and store in variable 'tags'
tags = BeautifulSoup(targetpage, 'html.parser')
```

### Read HTML tags

```python
# find results within table
table = tags.find('table', attrs={'class': 'tablesorter'})
results = table.find_all('tr')
print('Number of results', len(results))
```

### Add column name

```python
# First we will create the header for the dataset we will use to store the data

rows = []
rows.append(['Company', 'Price', '1-day Change', '52 Week Price', 'P/E', 'P/B', 'Enterprice Value(Cr)', 'Market Cap(Cr)'])
print(rows)
```

```python
# This function will take the company name tag and will extract company name and description.
def parseCompanyName(dataval) :
    # get the entire text in the tag 1 
    entiredata = dataval.getText()
    # now only fetch company name with tag 'span'
    companyname = dataval.find('span', attrs={'class':'company-name'}).getText()    
    # description is entire data excluding company name. So removing compnany name from entire data will give description.
    description = entiredata.replace(companyname, '')
    return [companyname, description]
```

If you see the url attached with the company name tag "http://www.fasttrack.co.uk/company_profile/plan-com" this is not the actual link to the company website. But its a link to another page that has an overview of each company. Each company page has it’s own table, which most of the time contains the company website as a bottom entry in a table.

To scrape the url from each table and save it as a variable, we need to use the same steps as above:

Find the element that has the url of the the company page on the fast track website
    
    Make a request to each company page url
    Parse the html using Beautifulsoup
    Find the elements of interest
    
 Looking at a few of the company pages, as in the screenshot above, the urls are in last row in the table so we can search within the last row for the <a> element.

```python
# will be used to extract the url from the company name field.
def getWebpage (dataval):
    # go to link and extract company website
    url = dataval.find('a').get('href')
    page = urllib.request.urlopen(url)
    # parse the html 
    soup = BeautifulSoup(page, 'html.parser')
    # find the last result in the table and get the link
    try:
        tableRow = soup.find('table').find_all('tr')[-1]
        webpage = tableRow.find('a').get('href')
    except:
        webpage = None
    return webpage
```

```python
# This function will parse the ease element and allocate the right value in the list format
def parsetag (data):
    # write columns to variables
    rank = data[0].getText()
    
    # get information about company name and description
    companyname = parseCompanyName(data[1])
    company = companyname[0]
    description = companyname[1]
    
    # get webpage
    webpage = getWebpage(data[1])
    
    location = data[2].getText()
    yearend = data[3].getText()
    salesrise = data[4].getText()
    sales = data[5].getText()
    
    # remove unwanted characters
    sales = sales.strip('*').strip('†').replace(',','')
    
    staff = data[6].getText()
    comments = data[7].getText()
    return [rank, company, webpage, description,location, yearend, salesrise,sales,staff,comments]
```
