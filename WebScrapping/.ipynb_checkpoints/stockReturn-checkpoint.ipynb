### Imports

```python
# beautifuSoup to handle the HTML data.
from bs4 import BeautifulSoup 

# To make connection to the web page
import urllib.request
```

```python
# specify the url
targeturl =  'https://www.valueresearchonline.com/stocks/stockselector/returns.asp?index=46'
```

```python
# query the website and return the html to the variable "targetpage"
targetpage = urllib.request.urlopen(targeturl)
# parse the html using beautiful soup and store in variable 'tags'
tags = BeautifulSoup(targetpage, 'html.parser')
```

### Read HTML tags

```python
# find results within table
table = tags.find('table', attrs={'class': 'tablesorter'})
results = table.find_all('tr')
print('Number of results', len(results))
```

```python
results[0:2]
```

### Add column name

```python
# First we will create the header for the dataset we will use to store the data

rows = []
rows.append(['Company', 'Price', '1-day Change', '52 Week Price', 'P/E', 'P/B', 'Enterprice Value(Cr)', 'Market Cap(Cr)'])
print(rows)
```

```python
# loop over results

for result in results:
    # find all columns per result
    data = result.find_all('td')
    # check that columns have data ; this will be true in case of first column which is hearder and do not have any td element
    # we also do not want to add this row.
    if len(data) == 0: 
         continue
     
    #get company name
    name = data[0].find('a').getText()
    print (name)
    
    #price
    entiredata = data[3].getText()
    date = data[3].find('font').getText()  
    price = entiredata.replace(date, '')
    print (price)
    
    #change
    change = data[4].getText()
    print(change)
    
    #52 week price
    week = data[5].getText()
    print (week)
    
    # P/E
    PE = data[6].getText()
    print (PE)
    
    #P/B
    PB = data[7].getText()
    print (PB)
    
    #EnterpriceVal
    EntVal = data[8].getText()
    print (EntVal)
    
    #marketCap
    marCap = data[9].getText()
    print (marCap)
 
  
    rows.append([name, price, change, week, PE, PB, EntVal, marCap])
    
```

### Write data to CSV

Following code will write the data into the company100.csv file

```python
import csv
# Create csv and write rows to output file
with open('overview100.csv','w', newline='') as f_output:
    csv_output = csv.writer(f_output)
    csv_output.writerows(rows)
```
